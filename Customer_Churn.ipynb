```python
# Replace blank spaces with '0' in the 'TotalCharges' column
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")

# Convert the 'TotalCharges' column to float type
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

```python
df.info()
```

```python
# Calculate the total number of missing values in the DataFrame
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
# Calculate the total number of duplicate customer IDs in the DataFrame
df["customerID"].duplicated().sum()
```

```python
# Convert 'SeniorCitizen' column from numerical (0,1) to categorical ('no', 'yes') to make it easier to understand
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
    
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
# Vizualize the count of customers by churn status with annotated bar label
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
# Plot a pie chart showing the percentage distribution of churned vs non churned
plt.figure(figsize=(3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

```python
# Visualizing customer churn distribution via gender using a count plot
plt.figure(figsize=(3, 3))
sns.countplot(x="gender", data=df, hue="Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
# Visualize the count of senior and non senior customers using a count plot
plt.figure(figsize=(4, 4))
ax = sns.countplot(x="SeniorCitizen", data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
# Create a stacked bar chart to visualize the percentage of churn by senior citizens
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))

total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Adding percentage labels on bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x+width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')
    
plt.title('Churn by Senior Citizen (Stacked Bar)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor=(0.9, 0.9))

plt.show()
```

```python
# Plot a histogram to compare tenure lengths between churned and non churned
plt.figure(figsize=(9,4))
sns.histplot(x="tenure", data=df,bins=72, hue="Churn")
plt.show()
```

```python
# The data shows that people who have used our service for a long time tend to stay. We see the most churn at 1-2 months.
```

```python
# Visualize customer churn distribution across diferrent contract types
plt.figure(figsize=(4, 4))
ax = sns.countplot(x="Contract", data=df, hue="Churn")
for container in ax.containers:
    ax.bar_label(container)
plt.title("Count of Customers by Contract")
plt.show()
```

```python
# Customers with yearly contracts are less likely to Churn.
```

```python
# Visualize the distribution of various services among customers, segmented by churn, using subplots
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# The majority of customers that did not churn had services like PhoneService, InternetService (DSL), and OnlineSecurity enabled. For the services: OnlineBackup, TechSupport, and StreamingTV, churn rates were higher.
```

```python
# Visualize customer churn distribution across different payment methods
plt.figure(figsize=(6, 4))
ax = sns.countplot(x="PaymentMethod", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation=45)
plt.show()
```

```python
# Customers are more likely to churn when using an electronic check as the payment method.
```
